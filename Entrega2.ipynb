```python
""" Programa main """
import random
import numpy as np

#class main
class Main: 
    listaPetroleros=[]
    listaRemolcadores =[]
    listEventos = []
    muelles = 0
    tiempo_max = 50
    sigmaVacio = 1
    muVacio = 2
    sigmaLleno = 3
    muLleno = 10
    grado = 2
    
    #init
    def __init__(self,remolcadores,muelles):
        # Lista de remolcadores
        self.listaPetroleros=ListPetroleros()
        self.listaRemolcadores = ListRemolcadores(remolcadores)
        self.listaEventos = ListEventos()
        self.muelles = muelles
    
    def movimiento_puerto_muelle(self, mu, sigma, tiempo):
        return tiempo + random.normalvariate(mu, sigma)
    
    def descarga(self, x, tiempo):
        return tiempo + (60 * np.random.chisquare(x))
    
    def muellesDisponibles(self):
        res = False
        if self.listaPetroleros.enMuelles() < self.muelles:
            res = True
        return res
    
    def between(self,value, min, max):
        if min <= value < max:
            return True
        else:
            return False
        
    def getPoissonRate(self,time):
        """ Método que devuelve el valor de la distribución de Poisson para un tiempo dado """
        lambd = 0
        h_in_day = (time / 60.0) % 24.0
        if self.between(h_in_day, 0.0, 5):
            lambd = 2.0 / 5.0 * h_in_day + 5.0
        elif self.between(h_in_day, 5, 8):
            lambd = -1.0 / 3.0 * h_in_day + 26.0 / 3.0
        elif self.between(h_in_day, 8, 15):
            lambd = 3.0 / 7.0 * h_in_day + 18.0 / 7.0
        elif self.between(h_in_day, 15, 17):
            lambd = -3.0 / 2.0 * h_in_day + 63.0/2.0
        elif self.between(h_in_day, 17, 24):
            lambd = -1.0 / 7.0 * h_in_day + 59.0 / 7.0
        else:
            print("Lambda fuera del rango")
            #logging.error("lambda out of index")
        return lambd

    # metodo que inicializa la llegada de todos los petroleros
    def init_tiempos_petroleros(self):
        """ Método que devuelve un array con 
        todos los tiempos de llegada de los petroleros """
        tiempo = 0
        cont=0
       

        # Calculamos el primero en llegar
        tiempo += 60 * random.expovariate(self.getPoissonRate(tiempo))
        self.listaPetroleros.añadirBarco(tiempo)
        self.listaEventos.añadirEvento("Petrolero",0,tiempo,0,-1)

        # Mientras las llegadas esten dentro del tiempo de simulacion calculamos las siguientes entradas
        while tiempo < self.tiempo_max:
            tiempo += 60 * random.expovariate(self.getPoissonRate(tiempo))
            self.listaPetroleros.añadirBarco(tiempo)
            cont=cont+1
            self.listaEventos.añadirEvento("Petrolero",cont,tiempo,0,-1)
        
    
    # funcion simular
    def simular(self):
        self.init_tiempos_petroleros()
        while self.listaPetroleros.final()==False:
            petrolero = self.listaPetroleros.list[0]
            #print(self.listaPetroleros.list[0])
           # print(self.listaPetroleros.list)
           # print(self.listaRemolcadores.list)
            if petrolero[2]==0 :
                #print("estado0")
                self.estado0(petrolero)
            elif petrolero[2]==1:
                #print("estado1")
                self.estado1(petrolero)
            elif petrolero[2]==2:
                #print("estado2")
                self.estado2(petrolero)
            
                
    def estado0(self,petrolero):
        if self.listaRemolcadores.colaEntrada(petrolero[1])==True and self.muellesDisponibles():
            #print("evento1")
            self.evento1(petrolero)
        elif self.listaRemolcadores.colaSalida(petrolero[1])==True and self.muellesDisponibles():
            #print("evento5")
            self.evento5(petrolero[1])#CAMBIAR EL PARAMETRO
            self.esperaEntrada(petrolero)
        else:
            
            if self.listaPetroleros.todosDesc(self.muelles)==True:
                #print("CuelloDeBotella")
                remolcador = self.listaRemolcadores.libreSalidaSinT()
                tiempo = self.listaPetroleros.getPetroleroDesc()
                self.listaRemolcadores.modificarAllRemolcadores(tiempo)
                self.esperaEntrada(petrolero)

            else:
                self.esperaEntrada(petrolero)
            #print("espera")
            
            
    def estado1(self,petrolero):
        #print("evento2")
        self.evento2(petrolero)
        
    
    def estado2(self,petrolero):
        if self.listaRemolcadores.colaSalida(petrolero[1])==True:
            #print("evento3")
            self.evento3(petrolero)
        elif self.listaRemolcadores.colaEntrada(petrolero[1])==True:
            #print("evento4")
            self.evento4(petrolero[1])
            self.esperaSalida(petrolero)
        else:
            #print("Esperar")
            self.esperaSalida(petrolero)
            
    def evento1(self,petrolero):
        tiempo = self.movimiento_puerto_muelle( self.muLleno, self.sigmaLleno,petrolero[1])
        remolcador = self.listaRemolcadores.libreEntrada(petrolero[1])
        
        self.listaEventos.añadirEvento("Remolcador",remolcador[0],tiempo,1,petrolero[0])
        self.listaEventos.añadirEvento("Petrolero",petrolero[0],tiempo,1,remolcador[0])
        
        self.listaRemolcadores.modificarRemolcador(remolcador[0],tiempo,1,petrolero[0])
        self.listaPetroleros.modificarPetrolero(tiempo,1,remolcador[0])
        
        
    def evento2(self,petrolero):
        tiempo = self.descarga( self.grado, petrolero[1])
        
        self.listaEventos.añadirEvento("Petrolero",petrolero[0],tiempo,2,-1)
        
        self.listaPetroleros.modificarPetrolero(tiempo,2,-1)
        
    def evento3(self,petrolero):
        tiempo = self.movimiento_puerto_muelle( self.muLleno, self.sigmaLleno, petrolero[1])
        remolcador = self.listaRemolcadores.libreSalida(petrolero[1])
        
        self.listaEventos.añadirEvento("Remolcador",remolcador[0],tiempo,0,petrolero[0])
        self.listaEventos.añadirEvento("Petrolero",petrolero[0],tiempo,3,remolcador[0])
        
        self.listaPetroleros.quitarPetrolero()
        self.listaRemolcadores.modificarRemolcador(remolcador[0],tiempo,0,petrolero[0])
        #self.listaPetroleros.modificarPetrolero(tiempo,3,remolcador[0])
        
                
    def evento4(self,tiempo):
        remolcador = self.listaRemolcadores.libreEntrada(tiempo)
        tiempo = self.movimiento_puerto_muelle( self.muVacio, self.sigmaVacio,remolcador[1])  
        
        self.listaEventos.añadirEvento("Remolcador",remolcador[0],tiempo,1,-1)
        
        self.listaRemolcadores.modificarRemolcador(remolcador[0],tiempo,1,-1)#FALLO CON ID Y POSICIóN      
                
    def evento5(self,tiempo):        
        remolcador = self.listaRemolcadores.libreSalida(tiempo)
        tiempo = self.movimiento_puerto_muelle( self.muVacio, self.sigmaVacio,remolcador[1])
        
        self.listaEventos.añadirEvento("Remolcador",remolcador[0],tiempo,0,-1)
        
        self.listaRemolcadores.modificarRemolcador(remolcador[0],tiempo,0,-1)#FALLO CON ID Y POSICIóN   
        
    def esperaEntrada(self,petrolero):
        remolcador = self.listaRemolcadores.list[0]
        tiempo = remolcador[1]
        self.listaPetroleros.modificarPetrolero(tiempo,0,-1)
        
    def esperaSalida(self,petrolero):
        remolcador = self.listaRemolcadores.list[0]
        tiempo = remolcador[1]
        self.listaPetroleros.modificarPetrolero(tiempo,1,-1)
        
       
```
